### <a name='0'></a> Import bibliotek

```
import numpy as np
import pandas as pd
import plotly.express as px
```

### <a name='1'></a> Wygenerowanie danych i wizualizacja

```
from sklearn.datasets import make_blobs

data = make_blobs(n_samples=1200, centers=3, cluster_std=1.2, random_state=42)[0]
df = pd.DataFrame(data, columns=['x', 'y'])
px.scatter(df, 'x', 'y', width=750, height=500, title='Klasteryzacja')
```

### <a name='2'></a> DBSCAN

```
from sklearn.cluster import DBSCAN

cluster = DBSCAN(eps=0.5, min_samples=5)
cluster.fit(data)
```

```
cluster.labels_[:10]
```

```
df['klaster']=cluster.labels_
df.head()
```

### <a name='3'></a> DBSCAN - wizualizacja

```
px.scatter(df, 'x', 'y', 'klaster', width=850, height=500, title='DBSCAN(eps=0.5, min_samples=5)')
```

```
cluster = DBSCAN(eps=0.5, min_samples=7)
cluster.fit(data)

df['cluster'] = cluster.labels_
px.scatter(df, 'x', 'y', 'cluster', width=850, height=500, title='DBSCAN(eps=0.5, min_samples=7)',
           )
```

```
cluster = DBSCAN(eps=0.8, min_samples=5)
cluster.fit(data)

df['cluster'] = cluster.labels_
px.scatter(df, 'x', 'y', 'cluster', width=950, height=500, title='DBSCAN(eps=0.7, min_samples=5)')
```
